```python
df = pd.read_csv('crime.csv')
print(df)
```

```python
plt.style.use('dark_background')
plt.rcParams['figure.figsize'] = (29,9)
sns.countplot(df['Population'], palette = 'gnuplot')
plt.title("Major a crime in india" ,  fontweight = 30, fontsize = 20)
plt.plot()
```

```python
import numpy as np
y = df['Murder'].value_counts().head(25)
    
plt.rcParams['figure.figsize'] = (15, 15)
plt.style.use('fivethirtyeight')

color = plt.cm.magma(np.linspace(0, 1, 15))
squarify.plot(sizes = y.values, label = y.index, alpha=.8, color = color)
plt.title('Tree Map for Top 25 Crimes', fontsize = 20)

plt.axis('off')
plt.show()
```

```python
from wordcloud import WordCloud
plt.rcParams['figure.figsize'] = (12,33)
plt.style.use('fast')
wc = WordCloud(background_color = 'orange', width = 1500, height = 1500).generate(str(df['Assault']))
plt.title('Description world year')
plt.imshow(wc)
plt.axis('off')

plt.show()

```
